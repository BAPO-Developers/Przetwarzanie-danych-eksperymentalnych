```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg as la
```

```python
N=100
x=np.linspace (0,5,N)
print (x.shape)
mu=0
sigma=0.1
szum=mu+sigma*np.random.randn(N)
y=np.log(x+1)+szum
plt.plot (x,y)
print (y.shape)
```

```python
A1=np.vstack([x**0, x**1]).T   #aproksymacja (f. jednej zmiennej) wielomianem I rzedu 
A2=np.vstack([x**0, x**1, x**2]).T   #jest do aproksymacji wielomianami II rzedu 
#trzeba dopisać A3 do aproksymacji f.jednej zmiennej) wielomianami III rzedu 
#oraz utworzyc A4 do aproksymacji f. dwóch zmiennych
#na lepsza ocene treba tez dla szeregu Fouriera (dla f. jednej zmiennej i dwoch zmiennych)
print (A1.shape)
print (A2.shape)

sol1, r1, rank1, sv1 = la.lstsq (A1,y)   #obliczenie wspolczynnikow - linia prosta
sol2, r2, rank2, sv2 = la.lstsq (A2,y)   #obliczenie wspolczynnikow wielomian II rzedu
y_fit1 = sol1[0] + sol1[1]*x
y_fit2 = sol2[0] + sol2[1]*x+ sol2[2]*x**2
fig,(ax1, ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.plot (x,y_fit1)
ax1.plot (x,y_fit2,'g')
print (r1, r2)   #opisane w dokumentacji la.lstsq
```

```python
def f(x, y):   #f.dwoch zmiennych z=f(x,y)
    return np.sin(np.log(x+2) * np.sqrt(y))   

N=200
x = 5 * np.random.rand (N)   #x z przedzialu [0,5)
y = 5 * np.random.rand (N)

z = f(x,y)  #obliczenie wartosci funkcji - znajomosc wartosci funkcji jest potrzebna 
            #do wyznaczenia wspolczynnikow za pomoca MNK (drugi argument la.lstsq )
    
    
```

```python
x_rys = np.linspace(0, 5, 30)   #do sporzadzenia rysunku 
y_rys = np.linspace(0, 5, 30)   #do sporzadzenia rysunku 
X, Y = np.meshgrid(x_rys, y_rys)
Z = f(X, Y)
#https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
X.shape
```
