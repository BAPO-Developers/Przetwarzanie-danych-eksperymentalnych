```python
import numpy as np
import pandas as pd
#from scipy import stats
import sklearn.linear_model as slm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.array(pd.read_csv('data.csv', header=None)).T
data = np.array([[data[0], data[1]], [data[2], data[3]], [data[4], data[5]], [data[6], data[7]]])
data
```

```python
regression = slm.LinearRegression()

for i, single_data in enumerate(data):
    print(f'ZBIÓR {i+1}')
    print(f'Wastość średnia x_{i+1}: {round(single_data[0].mean(), 2)}')
    print(f'Wariancja x_{i+1}: {single_data[0].var(ddof=1)}')
    
    print(f'Wastość średnia y_{i+1}: {round(single_data[1].mean(), 2)}')
    print(f'Wariancja y_{i+1}: {round(single_data[1].var(ddof=1), 2)}')
    
    print(f'Pearson {i+1}: {round(np.corrcoef(single_data[0], single_data[1])[0, 1], 3)}')
    regression.fit(single_data[0].reshape(-1, 1),single_data[1].reshape(-1, 1))
    b = round(regression.intercept_[0], 3)
    a = round(regression.coef_[0, 0],3)
    print(f'Przesunięcie: {b}')
    print(f'Współczynnik kierunkowy: {a}\n')
    
    plt.title(f'Zbiór nr {i+1}')
    plt.xlabel(f'$x_{i+1}$')
    plt.ylabel(f'$y_{i+1}$')
    plt.scatter(single_data[0], single_data[1], c='r')
    points = np.arange((min(single_data[0])*0.5), (max(single_data[0])*1.2), 0.001)
    plt.plot(points, points*a+b)
    plt.savefig(f'zbior_{i+1}.png')
    plt.show()
    
    print('\n\n\n')
```
